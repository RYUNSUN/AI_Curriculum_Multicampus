```python
import tensorflow as tf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
car_data=pd.read_csv("data/cars.csv")
```

```python
car_data.head()
```

```python
x=tf.placeholder(tf.float32,shape=[None])
y=tf.placeholder(tf.float32,shape=[None])
b=tf.Variable(tf.random_normal([1]))
w=tf.Variable(tf.random_normal([1]))

hf=w*x+b
loss=tf.reduce_mean(tf.square(hf-y))
train=tf.train.GradientDescentOptimizer(0.001).minimize(loss)
xtrain=car_data["mpg"].tolist()
ytrain=car_data["hp"].tolist()

sess=tf.Session()
sess.run(tf.global_variables_initializer())

for step in range(20001):
    sess.run(train, feed_dict={x:xtrain,y:ytrain})
    if step % 2000 == 0:
        wv, bv, lv = sess.run([w, b, loss], feed_dict={x:xtrain, y:ytrain})
        print("w:%s b:%s loss:%s" % (wv,bv,lv))
```
