```python
data=['A','B','C','D']
y_pos=np.arange(len(data))
perform=2+3*np.random.rand(len(data))
perform
plt.title('My Chart')
plt.xlabel("xlabel")
plt.barh(y_pos,perform,alpha=0.5,xerr=0.2) # xerr : 0.2의 오차범위를 나타냄
plt.yticks(y_pos,data) # y_pos(0,1,2,3) 각각에 대해서 data 값을 부여
plt.show()
```

```python
# stem plot : 그래프 폭이 없는 막대바 그래프 (이산확률을 나타낼때 종종 활용)
x=np.linspace(0.1, 2*np.pi,10) # 선형 공간 함수, 0.1 ~ 2*3.14 10개 데이터 생성
plt.stem(x,np.cos(x),"-.") # 선,점으로 구성
```

```python
labels=['Seoul','Busan','Jeju','KwangJu']
colors=['gold','silver','lightskyblue','blue']
sizes=[15,30,40,10] #합이 100(%)
explode=(0,0,0.1,0) # 0.3은 떼어냈을 때 간격
plt.pie(sizes,explode=explode,labels=labels,startangle=90,colors=colors,shadow=True,
       autopct='%1.1f%%') 
# startangle : 시각화를 시작하는 각도:90도, 12시를 기준으로 반시계방향
# explode :원하는 부분 강조 
# autopct : 각 데이터의 확률 값. 앞에 % 1개, 뒤 %%2개를 해줘야 확률값 1개만 들어감.
#           숫자자리확보. 소수 몇째 자리까지 출력할지
# matplotlib api 참고해서 수정할 것(https://matplotlib.org/api/pyplot_summary.html)
```

```python
# imshow : 2차원 이미지 데이터를 색상으로 표시
from sklearn.datasets import load_digits
digits=load_digits()
len(digits) #5
digits.images[0].shape #(8,8) : 64개의 픽셀로 구성
x=digits.images[3] # 0:흰색, 0이 아닌 수:검은색 으로 생각하면 숫자 3으로 보임
plt.imshow(x) # 컬러
plt.imshow(x,cmap=plt.cm.bone_r,interpolation='nearest') # cmap:그레이스케일로 변환
# interpolation : 색을 그라데이션으로 나타내 자연스럽게 표현
```

```python
# 등고선 그리기
def myFunc(x,y):
    return(1-x/2+x**4+y**2)*np.exp(-x**2-y**2) # 임의의 다항식  
x=np.linspace(-3,3,256)
y=np.linspace(-3,3,256)
xx,yy=np.meshgrid(x,y) # meshgrid:등고선을 그리고자하는 대상데이터를 나타내는 특별(?)한 행렬 
# => 외우려 하지말고 xx,yy 행렬 형태로 나타난다는 것만 알고 넘어가기
xx
yy
zz=myFunc(xx,yy)
zz
```

```python
plt.contourf(xx,yy,zz,alpha=0.75, cmap='jet') # 색상 변경
# plt.contour(xx,yy,zz)
```
