```python
#lambda 함수 
sel=lambda q:print(soup.select_one(q).string)
# 람다함수 이름:sel, 람다함수 인자:q
# 위 구문의 의미? sel 함수를 호출할 때 전돨되는 인자(선택자)
# 전달되는 인자로부터 soup에 저장된 html문서에서 해당 텍스트를 추출하여 출력
```

```python
# 태백산맥을 추출할 수 있는 여러 문법
# 태백산맥을 추출하고자 할때, b4는 유일하므로 id명만 쓰면됨
sel("#b4")
sel("li#b4") #li태그의 id가 b4라는 의미
sel("ul li#b4") #ul태그 밑에 li 태그의 id가 b4
sel("ul > li#b4") #ul태그 밑에 li 태그의 id가 b4
sel("#books #b4")
sel("#books > #b4")
sel("li[id='b4']") # 따옴표 기호를 서로 다르게 해줘야함
sel("li:nth-child(4)") # for문을 사용해 수치값 변경 시, li 태그 내용 모두 출력 가능
sel("li:nth-of-type(4)")
print("="*40)
soup.select("li") #list 구조
print(soup.select("li")[3].string)
print("="*40)
soup.find_all("li")[3].string
```

```python
# day11 misssion 3번
# https://stackoverflow.com/questions/tagged/python 질문, 답 추출
# 1번째 질문 copy selector : #question-summary-57067264 > div.summary > h3 > a
# 2번째 질문 copy selector : #question-summary-57067260 > div.summary > h3 > a
from urllib.request import urlopen
page=urlopen("https://stackoverflow.com/questions/tagged/python")
document=page.read()
soup=BeautifulSoup(document,'html.parser')
questions=soup.find(id="questions") # a태그에 해당되는 부분 출력
# questions.find("a") # a태그의 첫번째 내용 출력
# questions.find_all("a") # a태그로 묶여있는 영역 출력
questions_list=questions.find_all("a",class_="question-hyperlink")
for questions in questions_list:
    #print(questions.get_text()) #print(questions.string)
    #print(questions)
    # a 태그는 링크가 연결되어있어 클릭하면 연결된 페이지로 넘어간다는 의미 
    # href은 이동할 페이지 주소를 의미
    print("질문:", questions.string)
    print("="*40)
    print("답변:",'http://stackoverflow.com/' + questions.get('href'))
    print("="*40)
```
