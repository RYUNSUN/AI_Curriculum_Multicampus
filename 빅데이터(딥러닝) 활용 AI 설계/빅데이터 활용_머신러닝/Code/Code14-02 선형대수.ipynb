```python
from sklearn.datasets import load_digits
digits=load_digits()
samples=[0,10,20,30,1,11,21,31] #이미지번호
```

```python
digits.images # 숫자 이미지를 digits화
```

```python
d=[]
for i in range(8):
    d.append(digits.images[samples[i]]) #samples=[0,10,20,30,1,11,21,31] => 이미지번호
    # 마지막 숫자가 0으로 끝나는 숫자는 0, 1로 끝나는 숫자는 1
plt.figure(figsize=(8,2)) # 그림을 그리기 위한 윈도우, 8개 이미지를 8등분 시켜 출력하려고 함
for i in range(8):
    plt.subplot(1,8,i+1) # figure 하나를 여러개로 나눔, 1줄 8칸인 박스 8개 출력
    plt.imshow(d[i], cmap=plt.cm.bone_r, interpolation='nearest')
    plt.title("img {}".format(i+1))
    plt.grid(False)
    plt.xticks([])
    plt.yticks([])
plt.show()
```

```python
#2차원 이미지(8,8) -> 1차원 벡터(크기 64)
#d[0].shape #(8,8)
v=[]
for i in range(8):
    #print(d[i].reshape(64,1)) # 2차원 벡터화
    v.append(d[i].reshape(64,1))
    #print("-"*50)
#np.shape(v) #(8, 64, 1)
plt.figure(figsize=(8,2)) # 그림을 그리기 위한 윈도우, 8개 이미지를 8등분 시켜 출력하려고 함
for i in range(8):
    plt.subplot(1,8,i+1) # figure 하나를 여러개로 나눔, 1줄 8칸인 박스 8개 출력
    plt.imshow(v[i], aspect=0.2, cmap=plt.cm.bone_r, interpolation='nearest')
    plt.title("img {}".format(i+1))
    plt.grid(False)
    plt.xticks([])
    plt.yticks([])
plt.tight_layout(w_pad=3)    
plt.show()
# 이를 통해서 내적을 구해보면 각 이미지 숫자의 유사함를 알 수 있음
# 1. 연습문제
# 벡터의 내적을 수행 -> 가장 큰 내적을 출력 (이미지 번호 출력)
```

```python
from scipy import misc
img_rgb=misc.face() # 두더지 같은 이미지
img_rgb.shape #(768, 1024, 3) => (가로, 세로, 채널)

plt.subplot(221) # 2행 2열로 나누고 1은 맨 윗쪽 상단을 의미
plt.imshow(img_rgb[:,:,0], cmap=plt.cm.gray) # red 채널만 출력
plt.axis("off")
plt.title("red img")

plt.subplot(222) # 2행 2열로 나누고 1은 맨 오른쪽 상단을 의미
plt.imshow(img_rgb[:,:,1], cmap=plt.cm.gray) # green 채널만 출력
plt.axis("off")
plt.title("green img")

plt.subplot(223) # 2행 2열로 나누고 1은 맨 아래 왼쪽을 의미
plt.imshow(img_rgb[:,:,2], cmap=plt.cm.gray) # blue 채널만 출력
plt.axis("off")
plt.title("blue img")

plt.subplot(224) # 2행 2열로 나누고 1은 맨 아래 오른쪽을 의미
plt.imshow(img_rgb, cmap=plt.cm.gray)
plt.axis("off")
plt.title("rgb img")
```
