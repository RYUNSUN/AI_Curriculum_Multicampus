### 다중선형회귀(Multi-variable Linear Regression)

```python
# 모의고사 점수를 가지고 수능 예측
x1data=[70,90,88,95,75] #3월 모의고사
x2data=[80,88,90,98,65] #4월 모의고사
x3data=[75,93,90,100,70] #5월 모의고사
ydata=[150,185,180,200,140] # 수능점수
# 예측값 [155.54593 180.4961  182.21358 196.99005 141.24539]
```

```python
x1=tf.placeholder(tf.float32)
x2=tf.placeholder(tf.float32)
x3=tf.placeholder(tf.float32)
y=tf.placeholder(tf.float32)
w1=tf.Variable(tf.random_normal([1]), name="weigth1")
w2=tf.Variable(tf.random_normal([1]), name="weigth2")
w3=tf.Variable(tf.random_normal([1]), name="weigth3")
b=tf.Variable(tf.random_normal([1]), name="bias")
```

```python
hf=x1*w1+x2*w2+x3*w3+b
cost=tf.reduce_mean(tf.square(hf-y))
```

```python
train=tf.train.GradientDescentOptimizer(1e-5).minimize(cost)
sess=tf.Session()
init = tf.global_variables_initializer()
sess.run(init)
```

```python
for step in range(2001):
    cv, hv, _=sess.run([cost,hf,train], feed_dict={x1:x1data,x2:x2data,x3:x3data,y:ydata})
    if step % 10 == 0:
        print(step, "cost:", cv, "prediction:", hv)
```

### Queue Runners
Queue : 자료구조(입력 순서와 동일하게 출력됨)  
데이터가 너무 클때는 한꺼번에 읽기 부담스러워 Queue 구조로 데이터를 불러올 수 있음  
Queue Runners : 멀티스레드 기반 동작 => Coordinater  
스레드는 프로세스보다 더 작은 단위로 프로세스 내에서 독립적으로 동작해야하는 주체를 의미하고, 멀티스레드는 스레드가 동시에 동작하는 것이다.  

```python
QUEUE_LENGTH=20
q=tf.FIFOQueue(QUEUE_LENGTH, "float") # Queue를 만드는 함수, 
# 데이터 20개를 float 형식으로 저장할 수 있음
```

```python
enq_ops=q.enqueue_many(([1.0,2.0,3.0,4.0],)) # 큐에 데이터를 넣는 동작을 관리. 한번에 데이터 4개를 큐에다 집어넣으라고 정의
qr=tf.train.QueueRunner(q,[enq_ops,enq_ops,enq_ops]) #큐러너, 집어넣어야할 객체는 리스트로 넣어야함
# cpu 상황에 따라 처리되기 때문에 thread의 처리 순서는 알 수 없음
```

```python
sess=tf.Session()
coord=tf.train.Coordinator() # 큐러너에 있는 thread를 관리
threads=qr.create_threads(sess, coord=coord, start=True) # start=True로 줘야 동작이 됨

for step in range(10):
    print(sess.run(q.dequeue())) # dequeue:que에 저장되어있는 데이터를 꺼내오는 것 
    
coord.request_stop() # 스레드 정지
coord.join(threads) # 정지될때까지 기다리는 함수
```

```python
QUEUE_LENGTH=20 # 입출입 되는 데이터의 길이를 어떻게 줘야할지 고민해봐야함
q=tf.FIFOQueue(QUEUE_LENGTH, "float") 
enq_ops1=q.enqueue_many(([1.0,2.0,3.0,4.0],)) # 각 스레드
enq_ops2=q.enqueue_many(([5.0,6.0,7.0,8.0],))
enq_ops3=q.enqueue_many(([9.0,10.0,11.0,12.0],))
qr=tf.train.QueueRunner(q,[enq_ops1,enq_ops2,enq_ops3])
```

```python
sess=tf.Session()
coord=tf.train.Coordinator()
threads=qr.create_threads(sess, coord=coord, start=True) 

for step in range(30):
    print(sess.run(q.dequeue())) 
    
coord.request_stop() # 스레드 정지
coord.join(threads) # 정지될때까지 기다리는 함수
```

### 파일에서 데이터 읽기 작업 순서
1. 파일 목록 읽기
2. 파일 목록을 파일이름큐에 저장
3. reader가 파일이름큐에서 파일명을 하나씩 읽어온다
4. decoder는 파일에 있는 데이터를 읽어들인다
5. 데이터를 정제(전처리)
6. 정제된 데이터를 example queue에 저장
7. 학습 진행

```python
import tensorflow as tf
filename_queue= tf.train.string_input_producer(['q_1.txt','q_2.txt','q_3.txt'],shuffle=True)
reader = tf.TextLineReader()
key,value = reader.read(filename_queue)
a,b=tf.decode_csv(value, record_defaults=[[0],[0]],field_delim=',')
with tf.Session() as sess: 
    coord = tf.train.Coordinator()
    threads = tf.train.start_queue_runners(sess=sess, coord=coord)   
    for i in range(100):
        print(sess.run([a,b]))   
    coord.request_stop()
    coord.join(threads) 
```
